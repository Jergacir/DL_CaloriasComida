```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ANÃLISIS DE RESULTADOS - Modelo 1 y Modelo 2
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import json
import matplotlib.pyplot as plt
import numpy as np

# Cargar historial Modelo 2
with open('../resultados/modelo2/historial.json', 'r') as f:
    hist_m2 = json.load(f)

# AnÃ¡lisis Modelo 2
print("ğŸ“Š ANÃLISIS MODELO 2 - REGRESIÃ“N DE CALORÃAS")
print("="*70)

# Mejores Ã©pocas
best_epoch = hist_m2['val_mae'].index(min(hist_m2['val_mae'])) + 1
print(f"\nğŸ† Mejor Ã©poca: {best_epoch}")
print(f"   Val MAE: {min(hist_m2['val_mae']):.2f} kcal")

# EvoluciÃ³n
print(f"\nğŸ“ˆ EvoluciÃ³n:")
epocas_clave = [1, 10, 20, 30, 40, 50]
for ep in epocas_clave:
    if ep <= len(hist_m2['val_mae']):
        print(f"   Ã‰poca {ep:2d}: {hist_m2['val_mae'][ep-1]:.2f} kcal")

# GrÃ¡fica comparativa
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(hist_m2['train_mae'], label='Train')
plt.plot(hist_m2['val_mae'], label='Val')
plt.xlabel('Ã‰poca')
plt.ylabel('MAE (kcal)')
plt.title('Modelo 2: Error Absoluto Medio')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.hist(hist_m2['val_mae'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Val MAE (kcal)')
plt.ylabel('Frecuencia')
plt.title('DistribuciÃ³n de Errores')
plt.axvline(min(hist_m2['val_mae']), color='red', linestyle='--', 
            label=f'Mejor: {min(hist_m2["val_mae"]):.2f}')
plt.legend()

plt.tight_layout()
plt.savefig('../resultados/analisis_completo.png', dpi=150)
plt.show()

print("\nâœ“ GrÃ¡fica guardada: resultados/analisis_completo.png")
```
